#### **2. SCD - 2**

##### **Tạo bảng customer_scd_type_2 (Dữ liệu ngày 2023-10-17)**

```python
# Tạo một đối tượng cursor
cur = conn.cursor()

# Tạo bảng customer_scd_type_2
cur.execute("""
    CREATE TABLE IF NOT EXISTS s4_hauct.hauct_customer_scd_type_2 (
        cust_id VARCHAR(255),
        cust_nm VARCHAR(255),
        birth_date DATE,
        add_id VARCHAR(255),
        opn_dt DATE,
        end_dt DATE,
        record_created_time DATE DEFAULT '2023-10-17',
        record_updated_time DATE DEFAULT '3000-12-12',
        record_is_active INT DEFAULT 1,
        record_change_type VARCHAR(255) DEFAULT 'CDC_INSERT',
        PRIMARY KEY (cust_id, record_created_time)
    );
""")

# Sao chép dữ liệu ngày 2023-10-17 của bảng customer vào bảng customer_scd_type_2
cur.execute("""
    INSERT INTO s4_hauct.hauct_customer_scd_type_2 (cust_id, cust_nm, birth_date, add_id, opn_dt, end_dt)
    SELECT cust_id, cust_nm, birth_date, add_id, opn_dt, end_dt
    FROM ai4e_test.customer;
""")

cur.close()
```

##### **Chèn / Cập nhật dữ liệu ngày 2023-10-18 vào bảng customer_scd_type_2**

```python
# Tạo một đối tượng cursor
cur = conn.cursor()

# Tạo bảng tạm thời new_values là các bản ghi để thêm vào bảng customer_scd_type_2
cur.execute("""
    CREATE TEMP TABLE new_values AS
        SELECT cust_id, cust_nm, birth_date, add_id, opn_dt, end_dt, 
        '2023-10-18'::date as record_created_time, 
        '3000-12-12'::date as record_updated_time, 
        1 as record_is_active, 
        CASE
            WHEN EXISTS (SELECT 1 FROM s4_hauct.hauct_customer_scd_type_2 WHERE cust_id = customer_cdc_18.cust_id) THEN 'CDC_UPDATE'
            ELSE 'CDC_INSERT'
        END AS record_change_type
    FROM ai4e_test.customer_cdc_18
""")

# Cập nhật các bản ghi cũ (lấy cái cũ gần đây nhất)
cur.execute("""
    WITH max_date AS (
        SELECT cust_id, MAX(record_created_time) AS max_record_created_time
        FROM s4_hauct.hauct_customer_scd_type_2
        GROUP BY cust_id
    )
    UPDATE s4_hauct.hauct_customer_scd_type_2
    SET record_updated_time = new_values.record_created_time,
        record_is_active = 0
    FROM new_values, max_date
    WHERE s4_hauct.hauct_customer_scd_type_2.cust_id = new_values.cust_id 
    AND s4_hauct.hauct_customer_scd_type_2.record_created_time < new_values.record_created_time 
    AND s4_hauct.hauct_customer_scd_type_2.record_created_time = max_date.max_record_created_time
    AND s4_hauct.hauct_customer_scd_type_2.cust_id = max_date.cust_id;
""")

# Chèn các bản ghi mới
cur.execute("""
    INSERT INTO s4_hauct.hauct_customer_scd_type_2 (cust_id, cust_nm, birth_date, add_id, opn_dt, end_dt, record_created_time, record_updated_time, record_is_active, record_change_type)
    SELECT n.cust_id, n.cust_nm, n.birth_date, n.add_id, n.opn_dt, n.end_dt, n.record_created_time, n.record_updated_time, n.record_is_active, n.record_change_type
    FROM new_values n;
""")

# Xóa bảng tạm thời
cur.execute("DROP TABLE IF EXISTS new_values;")

# Đóng cursor
cur.close()
```
